# One Hot Encoding - Variables with Multiple Categories

```python
import pandas as pd
import numpy as np

data = pd.read_csv('mercedes.csv',usecols=['X1','X2','X3','X4','X5','X6'])
```

```python
data.head()
```

```python
#Print Unique Categories
print(data.nunique())

#for cols in data.columns:
 #   print(cols,':',len(data[cols].unique()),'labels')
```

```python
# Apply One Hot Encoding
pd.get_dummies(data, drop_first=True).shape
```

```python
data.head()
```

```python
# Take top 10 results
data.X2.value_counts().sort_values(ascending=False).head(10)
```

```python
# Lets get the top 10 most frequent categories of the variable
top_10 = [x for x in data.X2.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
data.head()
```

```python
for label in top_10:
    data[label]=np.where(data['X2']==label, 1, 0)
    
data[['X2']+top_10].head(10)
    
```
